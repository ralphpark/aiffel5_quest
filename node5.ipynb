# 5. 람다, 일급객체
5-2 람다 - 함수를 한줄로 간단하게 정의
5-3 일급객체 - 함수를 변수에 할당, 함수의 인자로 전달, 함수의 리턴값으로 리턴

1) 람다 표현식
람다를 통해 코드 한줄만으로 익명의 함수를 만들 수 있음
(lambda 매개변수 : 리턴값)(인수)

람다를 포함한 식을 람다 표현식이라고 함
--> 인수를 전달하는 코드를 쓸때 조금 더 효율적으로 작성할 수 있음
람다 표현식으로 쓴 함수는 람다함수

```python
# 일반함수와 람다 함수 비교 - 일반함수
def add_two(x):
    return x+2
print(add_two(5))
```

```python
# 일반함수와 람다 함수 비교 - 람다함수
(lambda x : x+2)(5)
```

람다 표현식 안에서는 변수를 만들 수 없음

lambda x : y = 10; x+y 와 같이 y변수를 추가하는 방식은 불가능

람다 표현식에서 바깥에 있는 변수 참조는 가능

y = 나폴리
(lambda x : x+y)(이탈리아) #인수를 x만 받기 때문에 x에 이탈리아 할당

2) 람다표현식을 인수로 사용하기

람다함수는 기본적으로 익명
--> 인수로 전달하는 코드를 쓸때 조금 더 효율적으로 작성 할 수 있음

람다함수를 다른 함수의 인수로 사용하는 함수 3가지
map(), filter(), reduce()

map() 함수
리스트나 튜플에 어떤 특별한 처리를 할 때 사용하는 함수
map(함수, 리스트 or 튜플)
map(적용시킬 함수, 적용할 값들)

```python
#일반함수
def second(x):
    return x*2
two = [i for i in range(10)]
map(second, two)
```

```python
# 파이썬 3부터 map()은 list 혹은 tuple 함수로 묶어야 리턴받을 수 있음
list(map(second, two))
```

```python
tuple(map(second, two))
```

```python
#리스트
two = [i for i in range(10)]
#map + 람다 함수
list(map(lambda x : x*2, two))
```

filter() 함수
리스트에 들어있는 원소들을 함수에 적용시켜서 결과가 참인 값들로 새로운 리스틀 만듦
filter(함수, 리스트 or 튜플)

```python
#일반함수
def odd(x):
    if x % 2 == 1:
        return x
#리스트
two = [i for i in range(10)]
#filter + 일반함수
list(filter(odd, two))
```

```python
#filter + 람다함수
list(filter(lambda x : x % 2 == 1, two))
```

reduce() 함수
이전 결과와 누적해서 반환하는 함수
reduce(함수, 이터러블객체)

```python
# reduce+ 일반함수
from functools import reduce
#일반함수
def f(x, y):
    return x+y
#리스트
a = [1,2,3,4,5]
#reduce(함수, 이터러블객체)
reduce(f, a)
```

```python
#reduce + 람다함수
a = [1,2,3,4,5]
reduce(lambda x,y : x+y, a)
```

```python
# Q. 리스트 컴프리헨션과 람다 표현식을 이용해 1부터 100까지의 합을 구하는 코드를 구현해보세요!
reduce(lambda x,y : x+y, [i for i in range(1,101)])
```

2. 일급객체
객체지향 프로그래밍에서 사용되는 개념 중 하나로 아래 조건을 만족하는 객체를 의미
 - 변수 혹은 데이터 구조(자료구조) 안에 그 객체를 담을 수 있다(할당할 수 있다)

 - 매개변수로 전달 할 수 이다.

 - 반환값(return value)으로 사용할 수 있다.

(이미 알고 있는 int, float, str 등의 타입은 일급 객체로 사용 가능)

파이썬에서는 함수도 일급객체이다.

 - 변수 혹은 데이터 구조(자료구조) 안에 그 객체를 담을 수 있다(할당할 수 있다)

--> 함수를 변수에 할당할 수 있다.

--> 함수를 리스트에 넣을 수 있다.

```python
# 함수를 변수에 할당
def add(a,b):
    return a+b
f1 = add
print(f1(1,2))
```

```python
# 함수를 리스트에 넣기
def add(a,b):
    return a+b
def minus(a,b):
    return a-b
calc = [add, minus]
for i in calc:
    print(i(10,5))
```

```python
# 함수를 매개변수로 전달 할 수 있다 (앞에서 배운 필터, 리듀스 맵 해당)
def odd(x):
    if x % 2 == 1:
        return x
two = [i for i in range(10)]
list(filter(odd, two))
```

```python
# 함수를 반환값으로 사용할 수 있다. 중첩함수 사용
def hello(name):
    def printer():
        print("hello", name)
    return printer

func = hello("john")
func()
```
